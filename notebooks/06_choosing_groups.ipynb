## Choosing groups for which we will generate signatures
Getting an idea what cell samples are in there - and what groups we should be able to identify in the cluster: 

We take the pragmatic approach to split the sample names at the dash

```python
primary_cells = !grep "primary cell" ../data/column_vars.processed.csv | cut -f5 -d, | tr -d '"' | sort | uniq -c 
primary_cells
```

```python
len(primary_cells)
```

more coarse: 

```python
cell_types = !grep "primary cell" ../data/column_vars.processed.csv | cut -f5 -d, | tr -d '"' | awk -F" - " '{print $1}' | sort | uniq -c 
cell_types
```

```python
len(cell_types)
```
